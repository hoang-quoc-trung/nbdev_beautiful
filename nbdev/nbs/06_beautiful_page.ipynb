## <font style="text-align: center; font-weight: bold; color: #7e5f17;">2. Website</font>

- <a class="link" href="https://quarto.org/docs/websites/">Quarto Websites</a> là một nền tảng website builder cung cấp các công cụ để tạo và quản lý website một cách dễ dàng và nhanh chóng. Người dùng có thể tạo ra các trang web chuyên nghiệp, thương mại điện tử và blog chỉ trong vài phút mà không cần có kiến thức về lập trình hoặc thiết kế web. Thật may mắn nbdev được xây dựng trên Quarto, do đó chúng ta có thể dễ dàng sử dụng Quarto trên nbdev.
- Một vài option có sẵn từ  để custom website navigation như: Hiển thị sidebar theo cấu trúc hierarchy, giao diện điều hướng page, sub-menus ...  
- Để bắt đầu, bạn hãy mở file config `_quarto.yml` trong folder `nbs` để có thể edit. 
 

### <font style="text-align: center; font-weight: bold;">Top Navigation </font>

- Để custom top navigation, chúng ta sẽ edit ở entry `navbar`. Ví dụ sau, bạn có thể chỉ định các file `.jpynb` hoặc một link website nào đó:

```python
#| output: false
#| eval: false
website:
  navbar:
    background: primary
    title: My site
    search: true
    left:
      - text: "Home"
        file: index.jpynb
      - talks.jpynb
      - about.jpynb 
```

- Tạo submenu:

```python
#| output: false
#| eval: false
left:
  - text: "More"
    menu:
      - talks.jpynb
      - about.jpynb 
```

::: {.callout-tip}
## Tip 

Chi tiết các option tham khảo  [here 📓](https://quarto.org/docs/websites/website-navigation.html#top-navigation)
:::

### <font style="text-align: center; font-weight: bold;">Side Navigation</font>

- Để custom side navigation, chúng ta sẽ edit ở entry `sidebar`.

```python
#| output: false
#| eval: false
website:
  sidebar:
    style: "docked"
    search: true
    contents:
      - section: "Basics"
        contents:
          - index.jpynb
          - knitr-basics.jpynb
          - jupyter-basics.jpynb
      - section: "Layout"
        contents:
          - markdown-layout.jpynb
          - knitr-layout.jpynb
          - jupyter-layout.jpynb
```

::: {.callout-tip}
## Tip 
Chi tiết các option tham khảo  [here 📓](https://quarto.org/docs/websites/website-navigation.html#side-navigation)
:::

### <font style="text-align: center; font-weight: bold;">Favicon</font>

- Hiển thị logo trên tab trình duyệt

```python
#| output: false
#| eval: false
website:
  favicon: logo.png
```

### <font style="text-align: center; font-weight: bold;">Page Footern</font>

- Custom nội dụng dưới cùng của page

```python
#| output: false
#| eval: false
website:
    page-footer: 
    left: "Copyright 2023, Trung Hoang" 
    right: 
      - icon: github
        href: https://github.com/hoangtrung020541/nbdev_beautiful
      - icon: twitter 
        href: https://twitter.com/ 
    background: Gray
    foreground: White
    border: false
```

### <font style="text-align: center; font-weight: bold;">Separators</font>

- Tách nội dung giữa các sidebar bằng `text: "---"`

```python
#| output: false
#| eval: false
website:
  sidebar:
    contents:
      - section: "Quick start"
        contents:
          - 03_how_to_use_nbdev.ipynb
          - 04_directives.ipynb
          - 05_create_page.ipynb
          - 06_beautiful_page.ipynb
      - text: "---"
      - section: "Examples"
        contents:
          - 00_card_overview.ipynb
          - 01_card.ipynb
          - 02_desk.ipynb
      - 07_test.ipynb
```

### <font style="text-align: center; font-weight: bold;">Sidebar Tools</font>

- Một vài tool nho nhỏ bao gồm icon và href trên sidebar, lấy icon ở đây <a class="link" href="https://icons.getbootstrap.com/">Bootstrap Icons</a>. 

```python
#| output: false
#| eval: false
website:
  sidebar:
    tools:
      - icon: facebook
        href: https://www.facebook.com/
      - icon: discord
        href: https://discord.com/
      - icon: twitter
        href: https://twitter.com
      - icon: github
        menu:
          - text: Source Code
            url:  https://github.com/hoangtrung020541/nbdev_beautiful
          - text: Report a Bug
            url:  https://github.com/hoangtrung020541/nbdev_beautiful/issues
```

### <font style="text-align: center; font-weight: bold;">Dark Mode</font>

- Bật chế độ cho phép chuyển đổi theme: darkly <-> flatly

```python
#| output: false
#| eval: false
format:
  html:
    theme:
      light: flatly
      dark: darkly
    css: styles.css
    toc: true
```

## <font style="text-align: center; font-weight: bold; color: #7e5f17;">3. Custom Content</font>

Trong nbdev, chúng ta có thể sử dụng  <a class="link" href="https://www.w3schools.com/html/default.asp">HTML</a>,  <a class="link" href="https://www.w3schools.com/css/default.asp">CSS</a>,  <a class="link" href="https://www.markdownguide.org/">Markdown</a> để decore nội dung của page thông qua `Markdown cell` trong notebook.

### <font style="text-align: center; font-weight: bold;">Formatting Links</font>

- [Test 1](07_test.ipynb)  --> `[Test 1](07_test.ipynb)`
- <a style="color: red; text-decoration: none;" href="07_test.ipynb">Test 2</a>  -->  `<a style="color: red; text-decoration: none;" href="07_test.ipynb">Test 2</a>`
- <a class="link-hover" href="07_test.ipynb">Test 3</a>  -->  `<a class="link-hover" href="07_test.ipynb">Test 3</a>`
    
::: {.callout-note}
    Test 1 sử dụng markdown.
    Test 2 sử dụng html.
	Đối với Test 3 hơi phức tạp, liên quan đến cả html và css nên cần phải sử dụng 
    file `styles.css` trong folder `nbs`. Cụ thể chúng ta sẽ tạo `class="link"` ở 
    đây, sau đó sẽ set các thuộc tính của class link tại file css như sau:
:::

```python
#| output: false
#| eval: false
.link{
    color: #73D0FF;
    text-decoration: none;
}
.link:hover{
    color: #686868;
    text-decoration: none;
}
.link:active{
    color: black;
    text-decoration: none;
}
```

### <font style="text-align: center; font-weight: bold;">Callout Blocks</font>


    
Để làm được như vậy, sử dụng  <a class="link" href="https://quarto.org/docs/authoring/callouts.html#markdown-syntax">markdown syntax</a> như sau:

```python
#| output: false
#| eval: false
::: {.callout-note}
    Note that there are five types of callouts, including:
    `note`, `warning`, `important`, `tip`, and `caution`.
:::
```

```python
#| output: false
#| eval: false
::: {.callout-tip}
    This is an example of a callout with a title.
:::
```

```python
#| output: false
#| eval: false
::: {.callout-caution collapse="true"}
    This is an example of a 'folded' caution callout that can be expanded by the user. You can 
    use `collapse="true"` to collapse it by default or `collapse="false"` to make a collapsible
    callout that is expanded by default.
:::
```
